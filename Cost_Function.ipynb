```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1.0 , 2.0 , 3.0 , 4.0])
y = np.array([200.0 , 250.0 , 300.0 , 350.0])

plt.scatter(x,y,color="r",marker="*")
plt.xlabel("Size")
plt.ylabel("Price")
plt.title("Housing Price")

plt.show()
```

```python
def compute_model(x,w,b):
    return w*x+b
```

```python
def compute_cost(x,y,w,b):

    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_w=w*x[i]+b
        cost_t=(f_w-y[i])**2
        cost=cost+cost_t

    final = cost/(2*m)
    return final
```

```python
w = int(input("Enter w: "))
b = int(input("Enter b: "))

costs = compute_cost(x,y,w,b)
print("Cost: ",costs)
```

```python
w_values = np.linspace(0,400,100)

cost_value = []

for w in w_values:
    cost_value.append(compute_cost(x,y,w,0))

plt.plot(w_values,cost_value)
plt.xlabel("w")
plt.ylabel("J(w)")
plt.title("Cost function")
plt.show()
```
